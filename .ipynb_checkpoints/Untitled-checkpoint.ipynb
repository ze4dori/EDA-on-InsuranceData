```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv('https://github.com/stedy/Machine-Learning-with-R-datasets/blob/master/insurance.csv?raw=true')
```

```python
plt.hist(data.charges)
plt.title("Distribution of Charges")
```

Необходимо спрогнозировать стоимость страховки, другими словами на выходе мы получим вещественное число. Следовательно перед нами задача регрессии.

## 1. Предобработка данных

### Логарифмирование

```python
data['charges'] = np.log(data['charges'])
plt.hist(data.charges)
plt.title("Distribution of Charges")
```

```python
data.head(3)
```

### Проверка и ликвидация (если есть) пропусков

```python
data.isnull().sum()
```

### Кодирование категориальных признаков

```python
from sklearn.preprocessing import LabelEncoder

data['sex'] = LabelEncoder().fit_transform(data['sex'])
data['smoker'] = LabelEncoder().fit_transform(data['smoker'])
#data['region'] = LabelEncoder().fit_transform(data['region'])
```

```python
data_region = pd.get_dummies(data['region']).astype(int)
data = pd.concat([data, data_region], axis=1)
```

```python
data.head(3)
```

### Разделение выборки на тренировочную и тестовую

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['charges', 'region'], axis=1)
y = data['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
X_train.head(3)
```

### Нормирование численных переменных

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 2. Обучение на тренировочном множестве

### Линейная модель

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
df_coef = pd.DataFrame(lin_reg.coef_, X.columns, columns=['Coef'])
df_coef.T
```

```python
lr_r2_score = lin_reg.score(X_test, y_test)
lr_r2_score
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
palette = sns.color_palette('Set2')
sns.set_palette(palette)

plt.plot(y_test,y_pred,'o')
a, b = np.polyfit(y_test,y_pred, 1)
plt.plot(y_test, a*y_test + b)

plt.xlabel('True values')
plt.ylabel('Predicted values')

plt.show()
```

```python
sns.scatterplot(x=X_test[:, 2], y=y_pred, label='pred')
sns.scatterplot(x=X_test[:, 2], y=y_test, label='true')

plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()

plt.show()
```

### Деревянная модель

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=6, random_state=0)
tree_reg.fit(X_train, y_train)
```

```python
tr_r2_score = tree_reg.score(X_test, y_test)
tr_r2_score
```

```python
y_pred = tree_reg.predict(X_test)
```

```python
sns.set_palette(palette)
plt.plot(y_test,y_pred,'o')
a, b = np.polyfit(y_test,y_pred, 1)
plt.plot(y_test, a*y_test + b)

plt.xlabel('True values')
plt.ylabel('Predicted values')

plt.show()
```

```python
sns.scatterplot(x=X_test[:, 2], y=y_pred, label='pred')
sns.scatterplot(x=X_test[:, 2], y=y_test, label='true')

plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()

plt.show()
```

### K-ближайших соседей

```python
from sklearn.neighbors import KNeighborsRegressor

kn_reg = KNeighborsRegressor(n_neighbors=8)
kn_reg.fit(X_train, y_train)
```

```python
kn_r2_score = kn_reg.score(X_test, y_test)
kn_r2_score
```

```python
y_pred = kn_reg.predict(X_test)
```

```python
sns.set_palette(palette)
plt.plot(y_test,y_pred,'o')
a, b = np.polyfit(y_test,y_pred, 1)
plt.plot(y_test, a*y_test + b)

plt.xlabel('True values')
plt.ylabel('Predicted values')

plt.show()
```

```python
sns.scatterplot(x=X_test[:, 2], y=y_pred, label='pred')
sns.scatterplot(x=X_test[:, 2], y=y_test, label='true')

plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()

plt.show()
```

### Случайный лес

```python
from sklearn.ensemble import RandomForestRegressor

rand_forest = RandomForestRegressor(n_estimators = 16, random_state = 0)
rand_forest.fit(X_train, y_train)
```

```python
rf_r2_score = rand_forest.score(X_test, y_test)
rf_r2_score
```

```python
y_pred = rand_forest.predict(X_test)
```

```python
sns.set_palette(palette)
plt.plot(y_test,y_pred,'o')
a, b = np.polyfit(y_test,y_pred, 1)
plt.plot(y_test, a*y_test + b)

plt.xlabel('True values')
plt.ylabel('Predicted values')

plt.show()
```

## 4. Сравнение метрик

Метод Случайный лес показал лучший результат среди всех (0.724).

```python
df = pd.DataFrame({'Sample': columns,
                'MAE': lr_mae,
                'MSE': lr_mse, 
                'RMSE': lr_rmse, 
                'MAPE': lr_mape})
lr_metrics = df.set_index('Sample')
lr_metrics
```

Получилась не нулевая метрика MAE равная 0.27, значит, что предсказанные значения близки к истинным значениям. На тестовой выборке значения также недалеки от нуля, что указывает на то, что модель хорошо обучается на новых данных, а не переобучается на тренировочных.

```python
df = pd.DataFrame({'Sample': columns,
                'MAE': tr_mae,
                'MSE': tr_mse, 
                'RMSE': tr_rmse, 
                'MAPE': tr_mape})
tr_metrics = df.set_index('Sample')
tr_metrics
```

Средняя абсолютная ошибка (MAE) в процессе обучения равна 0.17, в то время как на тестовом наборе она равна 0.25.

```python
df = pd.DataFrame({'Sample': columns,
                'MAE': kn_mae,
                'MSE': kn_mse, 
                'RMSE': kn_rmse, 
                'MAPE': kn_mape})
kn_metrics = df.set_index('Sample')
kn_metrics
```

По метрике MAE видно, что результаты предсказаний недалеки от истинных значений целевой переменной, при этом разница между метриками не велика, чтобы говорить о переобучении.

```python
df = pd.DataFrame({'Sample': columns,
                'MAE': rf_mae,
                'MSE': rf_mse, 
                'RMSE': rf_rmse, 
                'MAPE': rf_mape})
rf_metrics = df.set_index('Sample')
rf_metrics
```

Аналогично предыдущей модели разница между метриками MAE для тестовой и тренировочной выборке не слишком значительна, что говорит о том, что модель может все еще хорошо обобщать свои знания на новые данные.

В результате получаем, что все модели продемонстрировали нормальное обучение.
